```python
# Data Analysis using Pandas and Matplotlib

This project demonstrates loading a CSV file, performing basic data analysis
using Pandas, and visualizing insights using Matplotlib (bar chart, scatter plot,
and heatmap).
```

```python
import pandas as pd

# Create sample data
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Age": [23, 25, 30, 35, 40],
    "Salary": [30000, 35000, 50000, 65000, 80000],
    "Experience": [1, 2, 5, 8, 12]
}

# Create DataFrame
df = pd.DataFrame(data)

# Save as CSV file
df.to_csv("data.csv", index=False)

df
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("data.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
avg_salary = df["Salary"].mean()
avg_age = df["Age"].mean()

print("Average Salary:", avg_salary)
print("Average Age:", avg_age)
```

```python
plt.bar(df["Name"], df["Salary"])
plt.xlabel("Name")
plt.ylabel("Salary")
plt.title("Salary Distribution")
plt.show()
```

```python
plt.scatter(df["Experience"], df["Salary"])
plt.xlabel("Experience (Years)")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
corr
```

```python
plt.imshow(corr, cmap="coolwarm")
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("data.csv")

print(df.head())
print(df.describe())

print("Average Salary:", df["Salary"].mean())
print("Average Age:", df["Age"].mean())

plt.bar(df["Name"], df["Salary"])
plt.title("Salary Distribution")
plt.show()

plt.scatter(df["Experience"], df["Salary"])
plt.title("Experience vs Salary")
plt.show()

corr = df.corr(numeric_only=True)
plt.imshow(corr, cmap="coolwarm")
plt.colorbar()
plt.title("Correlation Heatmap")
plt.show()
```
